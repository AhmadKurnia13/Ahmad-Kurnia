```python
import numpy as np

print("\nNama: Ahmad Kurnia")
print("NPM: 24552011297")
print("Kelas: TIF RP 24A\n")

#soal no 1
# Definisi turunan-turunan dari y' = x^2 - y
def f(x, y):
    return x**2 - y

def f_prime(x, y):
    return 2 * x - x**2 + y

def f_double_prime(x, y):
    return 2 - 2 * x + x**2 - y

def f_triple_prime(x, y):
    return -2 + 2 * x - x**2 + y

# === Fungsi metode Taylor orde 4 ===
def taylor_series_order_4_with_table(x0, y0, h, target_x, tolerance):
    x = x0
    y = y0
    i = 0

    # Cetak header tabel
    print("Soal 1. Metode Deret Taylor Orde 4")
    print("-" * 60)
    print(f"{'i':<3} {'x_i':<10} {'y(x_i)':<15} {'delta_y':<15}")
    print("-" * 60)

    # Cetak kondisi awal
    print(f"{i:<3} {x:<10.4f} {y:<15.6f} {'-':<15}")

    while abs(x - target_x) > tolerance:
        # Menghitung kontribusi delta_y menggunakan deret Taylor orde 4
        delta_y = h * f(x, y) + \
                  (h**2 / 2) * f_prime(x, y) + \
                  (h**3 / 6) * f_double_prime(x, y) + \
                  (h**4 / 24) * f_triple_prime(x, y)

        y_next = y + delta_y

        x += h
        y = y_next
        i += 1

        # Cetak hasil iterasi saat ini
        print(f"{i:<3} {x:<10.4f} {y:<15.6f} {delta_y:<15.6f}")

    print("-" * 60)
    print(f"{'h =':<6} {h:.6f}")
    print(f"{'Hasil y(0.5) ≈':<20} {y:.6f}")
    print("\n")

    return y

# --- Parameter dan Eksekusi ---
# Parameter soal
x0 = 0.0
y0 = 1.0
h = 0.1
target_x = 0.5
tolerance = 1e-5

# Jalankan fungsi
y_at_0_5 = taylor_series_order_4_with_table(x0, y0, h, target_x, tolerance)


# Soal nomor 2

def f(x):
    return (x**2 / 2) + (1 / (2 * x**2))  # Perbaikan: tambahkan tanda kurung untuk penyebut

def simpson_three_eighth(a, b, n):
    if n % 3 != 0:
        raise ValueError("n harus kelipatan 3 untuk Simpson 3/8")

    h = (b - a) / n
    hasil = f(a) + f(b)  # Inisialisasi dengan nilai di ujung interval
    print("Soal 2: Metode Simpson 3/8")
    print("-" * 60)
    print(f"{'i':<3} {'x_i':<10} {'f(x_i)':<15} {'Bobot':<10} {'Kontribusi':<15}")
    print("-" * 60)

    for i in range(1, n):
        x_i = a + i * h
        fx = f(x_i)

        if i % 3 == 0:  # Titik kelipatan 3 bobotnya 2
            bobot = 2
        else:            # Titik lainnya bobot 3
            bobot = 3

        kontribusi = bobot * fx
        hasil += kontribusi

        print(f"{i:<3} {x_i:<10.4f} {fx:<15.6f} {bobot:<10} {kontribusi:<15.6f}")

    integral = (3 * h / 8) * hasil
    print("-" * 60)
    print(f"{'h =':<6} {h:.6f}")
    print(f"{'Total =':<10} {hasil:.6f}")
    print(f"{'Hasil Integral ≈':<20} {integral:.6f}")
    return integral

a = 1/2
b = 2
n = 12
print(f"\nHasil pendekatan integral menggunakan Simpson 3/8 adalah: {simpson_three_eighth(a, b, n):.6f}")
```
